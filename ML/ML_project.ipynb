# Интернет-магазин «В один клик»

**Цели**:

(1) Разработка модели, предсказывающей вероятность снижения покупательской активности

(2) Формулировка предложений по повышению покупательской активности покупателей

### Описание данных

<div style="border:solid orange 2px; padding: 20px">
    
#### market_file
*данные о поведении покупателя на сайте, о коммуникациях с покупателем и его продуктовом поведении*

   - **id** (номер покупателя в корпоративной базе данных)
   - **Покупательская активность** (рассчитанный класс покупательской активности (целевой признак): «снизилась» или «прежний уровень»)
   - **Тип сервиса** (уровень сервиса, например «премиум» и «стандарт»)
   - **Разрешить сообщать** (информация о том, можно ли присылать покупателю дополнительные предложения о товаре. Согласие на это даёт покупатель.)
   - **Маркет_актив_6_мес** (среднемесячное значение маркетинговых коммуникаций компании, которое приходилось на покупателя за последние 6 месяцев. Это значение показывает, какое число рассылок, звонков, показов рекламы и прочего приходилось на клиента.)
   - **Маркет_актив_тек_мес** (количество маркетинговых коммуникаций в текущем месяце.)
   - **Длительность** (значение, которое показывает, сколько дней прошло с момента регистрации покупателя на сайте.)
   - **Акционные_покупки** (среднемесячная доля покупок по акции от общего числа покупок за последние 6 месяцев)
   - **Популярная_категория** (самая популярная категория товаров у покупателя за последние 6 месяцев)
   - **Средний_просмотр_категорий_за_визит** (показывает, сколько в среднем категорий покупатель просмотрел за визит в течение последнего месяца)
   - **Неоплаченные_продукты_штук_квартал** (общее число неоплаченных товаров в корзине за последние 3 месяца)
   - **Ошибка_сервиса** (число сбоев, которые коснулись покупателя во время посещения сайта)
   - **Страниц_за_визит** (среднее количество страниц, которые просмотрел покупатель за один визит на сайт за последние 3 месяца)

<div style="border:solid lightgreen 2px; padding: 20px">
    
#### market_money
*данные о выручке, которую получает магазин с покупателя, то есть сколько покупатель всего потратил за период взаимодействия с сайтом*

   - **id** (номер покупателя в корпоративной базе данных)
   - **Период** (название периода, во время которого зафиксирована выручка. Например, 'текущий_месяц' или 'предыдущий_месяц')
   - **Выручка** (сумма выручки за период)


<div style="border:solid blue 2px; padding: 20px">
    
#### market_time
*данныe о времени (в минутах), которое покупатель провёл на сайте в течение периода*

   - **id** (номер покупателя в корпоративной базе данных)
   - **Период** (название периода, во время которого зафиксировано общее время)
   - **минут** (значение времени, проведённого на сайте, в минутах)


<div style="border:solid yellow 2px; padding: 20px">
    
#### money
*данные о среднемесячной прибыли продавца за последние 3 месяца: какую прибыль получает магазин от продаж каждому покупателю*

   - **id** (номер покупателя в корпоративной базе данных)
   - **Прибыль** (значение прибыли)

````python
def predict_wrapper(X):
    if isinstance(X, pd.DataFrame):
        # Ensure DataFrame has correct original columns
        return best_pipe.predict_proba(X[num_columns + ohe_columns])
    else:
        # For numpy arrays, use the preprocessor's input shape
        if X.shape[1] == len(num_columns + ohe_columns):
            # Raw features before encoding
            X_df = pd.DataFrame(X, columns=num_columns + ohe_columns)
            return best_pipe.predict_proba(X_df)
        elif X.shape[1] == len(true_feature_names):
            # Already transformed features
            return best_pipe.named_steps['models'].predict_proba(X)
        else:
            raise ValueError(f"Input has {X.shape[1]} features, expected {len(num_columns + ohe_columns)} or {len(true_feature_names)}")
````

````python
def get_feature_names(preprocessor):
    feature_names = []
    for name, transformer, features in preprocessor.transformers_:
        if name == 'num':
            if transformer == 'passthrough':
                feature_names.extend(features)
            else:
                feature_names.extend(features)  # Numerical features keep original names
        elif name == 'ohe':
            if hasattr(transformer, 'get_feature_names_out'):
                # Get one-hot encoded feature names
                ohe_features = transformer.get_feature_names_out(features)
                feature_names.extend(ohe_features)
            else:
                feature_names.extend(features)
    return feature_names
````

   - удален выброс в переменной выручки
   - сформирован датасет clients с информацией о пользователях, которые были активны в течение трех месяцев
  
3. Исследовательский анализ данных
   - Покупательская активность 40% покупателей снизилась, активность оставшихся 60% осталась на прежнем уровне. Что указывает на то, что у сервиса действительно есть проблема с удержанием уровня покупательской активности
   - Около 70% пользователей сервиса используют стандарт тариф и 30% пользователей - премиум
   - Более 70% покупателей дали согласие на получение дополнительных предложений о товарах
   - Наиболее распространенной популярной категорией покупок среди пользователей являются товары для детей (25%), домашний текстиль (20%), косметика и аксессуары (17%), красота и здоровье (14%), мелкая бытовая техника и электроника (13%), кухонная посуда (10%)
   - Распределение прибыли продавца с каждого покупятеля также близко к нормальному. 50% покупателей приносят по 4045 прибыли ежемесячно.
  
4. Для поиска наилучшей модели был разработан пайплайн, который перебирает несколько моделей и их гиперпараметров, а также разные метода масштабирования данных

5. В результате анализа значимых признаков:
   - увеличение доли акционных покупок и количество неоплаченных продуктов повышают вероятность снижения покупательской активности
   - увеличение показателей активностей на сайте, включая количество просмотренных страниц за визит,количество минут, проведенных на сайте в текущем и предыдущем месяце, среднее количество просмотренных категорий за визит и маркетинговая активность поддерживают покупательскую способность покупателей

6. Была произведена сегментция покупателей на группы раска снижения покупательской активности. Исходя из наиболее значимых характеристик был разработа балл, которых представляет из себя среднее взвешенное наиболее значимых критериев.
   - при расчете балла учитывались следующие критерии (вес был выбран на основании shap значений):
       * акционные_покупки > 0.3 (0.4)
       * страниц_за_визит < 5 (0.12)
       * текущий_месяц_минут < 10 (0.12)
       * предыдущий_месяц_минут < 10.75 (0.12)
       * средний_просмотр_категорий_за_визит < 2 (0.12)
       * маркет_актив_6_мес < 3.7 (0.12)
    
   - на основании segment_score покупатели были разделены на три группы риска снижения покупательской активности:
       * низкий (segment_score < 0.12) - 835 пользователей
       * средний (0.12 < segment_score < 0.4) - 123 пользователей
       * высокий (segment_score >= 0.4) - 338 пользователей
